## Salary comparison between highest paying companies for Data Scientist role from Ambitionbox

[Link](https://www.ambitionbox.com/profile/data-scientist-salary)

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = "https://www.ambitionbox.com/profile/data-scientist-salary"
```

```python
request = requests.get(url)
request
```

```python
soup =  BeautifulSoup(request.text,'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
job = soup.find_all("div",{'class':'result-row'})
```

```python
company1 = job[0]
company1
```

### getting data of 1 company

```python
name = company1.find("div",{'class':'name'})
#c=re.findall('([a-zA-Z]*)[^\n\t]*',name)[0]
#for data in name:
#    name=data.get_text().split()[0]
#    print(name)
name=" ".join([data.get_text().split()[0] for data in name]) # " ".join() is used to convert from list to string
name
```

```python
exp = company1.find("div",{'class':'sbold-list-header'}).find({'span':'data-v-b24b8478'})
#exp =re.findall('([0-9-]+)',exp)[0]
for data in exp:
    print(data.get_text().split()[0])
```

```python

salary = company1.find("p",{'class':'averageCtc'}).text.strip()
salary
```

### getting all companies 

```python
company_list = []
for i in job:
    temp_comp = {}
    company_name = i.find("div",{'class':'name'})
    company_name = " ".join([data.get_text().split()[0] for data in company_name])
    temp_comp['Company name'] = company_name
    
    experience = i.find("div",{'class':'sbold-list-header'}).find({'span':'data-v-b24b8478'})
    experience = " ".join([data.get_text().split()[0] for data in experience])
    temp_comp['Yrs of experience'] = experience
    
    temp_comp['Avg salary'] = i.find("p",{'class':'averageCtc'}).text.strip()
    company_list.append(temp_comp)
```

```python
company_list
```

```python
df = pd.DataFrame(company_list)
df
```
